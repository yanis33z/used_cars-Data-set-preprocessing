# Estimation d'un AR(1) par Descente de Gradient

### Importation de pandas

```python
import pandas as pd
```

### Importation de la base de données

```python
df=pd.read_csv('../Data.csv',sep=';')
df.head(10)
```

### Définition des variables du modèle

```python
X_t=df['USA'][1:]
X_ret=df['USA'][:-1]

print(X_t,X_ret)
```

### Initialisation des paramètres

```python
rho=0
max_rep=500
k=0
eta_i=0.1
eta_f=0.0001
eps=10**(-6)
T=len(df['USA'])-1
```

### Estimation par descente de gradient

```python
while k<max_rep:
    eta=eta_i-(eta_i-eta_f)*(k/max_rep) #Linear dicreasing learning rate
    grad=0
    for t in range (T):
        grad+=(1/T)*X_ret.iloc[t]*(rho*X_ret.iloc[t]-X_t.iloc[t])
    n_rho=rho-eta*grad
    if abs(n_rho-rho)<eps:
        break
    else:
        rho=n_rho
        k+=1
print(rho,k)
```
